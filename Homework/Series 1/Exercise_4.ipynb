We want to generate a control input will move the drone towards the point $(3,3)$ starting from $(0,0)$. In order to
do so, we define a cost that penalizes the distance to the goal at each time step while minimizing velocities and the amount of trust needed,
i.e. we want to solve the following optimal control problem
$$\begin{align}
&\min_{x_0, u_0, x_1, u_1, \cdots} \frac{1}{2}\sum_{n=0}^{N} (x_n - x_{desired})^T Q (x_n - x_{desired}) + u_n^T R u_n\\
\textrm{subject to}\ \ & x_{n+1} = A x_{n} + B u_n\\
& x_0 = [0,0,0,0,0,0]^T
\end{align}$$
where $x_{desired} = [3,0,3,0,0,0]^T$

2. Write code to solve the problem for N=500 time steps (you will need to find diagonal matrices $Q>0$ and $R>0$ to create a nice movement) by solving the KKT system of linear equations. Do NOT inverse the KKT matrix, instead use the [NumPy solve function](https://numpy.org/doc/stable/reference/generated/numpy.linalg.solve.html) which is numerically more stable and efficient.
4. Show plots of all the states of the robot as a function of time
5. Show plots of the optimal control as a function of time

## 1. KKT Conditions

The KKT conditions for this optimal control problem are:

1. **Primal Feasibility**:
   $$x_{n+1} = Ax_n + Bu_n, \quad n = 0, \ldots, N-1$$

2. **Dual Feasibility**:
   $$\lambda_n = Q(x_n - x_{\text{desired}}) + A^T\lambda_{n+1}, \quad n = N-1, \ldots, 0$$
   $$\lambda_N = Q(x_N - x_{\text{desired}})$$

3. **Stationarity**:
   $$Ru_n + B^T\lambda_{n+1} = 0, \quad n = 0, \ldots, N-1$$

````python
# Initialization of the time steps, A & B matrix and the desired value of x
N = 500
A = quadrotor.A
B = quadrotor.B
goal = np.array([3, 0, 3, 0, 0, 0]).reshape(-1, 1)

# Diagonal matrices Q and R, which will work as a cost function
Q = np.diag([8, 2, 9, 3, 1, 4])
R = np.diag([0.15, 0.15])

# KKT matrix initialization for the problem
n_x, n_u = 6, 2
KKT = np.zeros(((N+1)*n_x + N*n_u, (N+1)*n_x + N*n_u))

# Constructing the KKT matrix as per the constraints above 
for i in range(N):  
    KKT[i*n_x:(i+1)*n_x, i*n_x:(i+1)*n_x] = Q
    
    if i < N:
        KKT[i*n_x:(i+1)*n_x, (N+1)*n_x+(i*n_u):((N+1)*n_x+(i+1)*n_u)] = B
        KKT[(N+1)*n_x+(i*n_u):((N+1)*n_x+(i+1)*n_u), i*n_x:(i+1)*n_x] = B.T
        KKT[(N+1)*n_x+(i*n_u):((N+1)*n_x+(i+1)*n_u), (N+1)*n_x+(i*n_u):((N+1)*n_x+(i+1)*n_u)] = R
        KKT[(i+1)*n_x:(i+2)*n_x, i*n_x:(i+1)*n_x] = -A
        KKT[i*n_x:(i+1)*n_x, (i+1)*n_x:(i+2)*n_x] = -np.eye(n_x)

# Initial condition of the KKT matrix
KKT[:n_x, :n_x] += np.eye(n_x)

# RHS vector
rhs = np.zeros(((N+1)*n_x + N*n_u, 1))
for i in range(N+1):
    rhs[i*n_x:(i+1)*n_x] = Q @ goal

# final solution of the KKT matrix using the NumPy solve funtion
sol = np.linalg.solve(KKT, rhs)
print(sol)
````

````python
# States and Controls of the robot to find out the plots
x_opt = sol[:(N+1)*n_x].reshape(N+1, n_x)
u_opt = sol[(N+1)*n_x:].reshape(N, n_u)
````

````python
# Plots of all the states of the robot as a function of time
plt.figure(figsize=(12, 8))
labels = ['x', 'x_dot', 'y', 'y_dot', 'theta', 'theta_dot']
for i in range(n_x):
    plt.subplot(3, 2, i+1)
    plt.plot(x_opt[:, i])
    plt.title(f'{labels[i]} over time')
plt.show()
````

````python
# Plots of the optimal control as a function of time
plt.figure(figsize=(10, 5))
for i in range(n_u):
    plt.subplot(1, 2, i+1)
    plt.plot(u_opt[:, i])
    plt.title(f'u{i+1} over time')
plt.show()
````
