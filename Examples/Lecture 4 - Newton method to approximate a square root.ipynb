Let's use Newton method to approxime $\sqrt{2}$. To do so, we search for the zero of the function  $ f(x) = x^2 - 2$

```python
import matplotlib.pyplot as plt
import numpy as np

def f(x):
  return  (x**2 - 2)

def f_prime(x):
  return 2*x

# Parameters
tolerance = 1e-10
max_iterations = 1000
x = 2  # Initial guess
iterations = []
approximations = []


for i in range(max_iterations):
    # Store the current iteration and approximation
    iterations.append(i)
    approximations.append(x)

    # Newton step
    x_new = x -f(x) / f_prime(x)

    # Check stopping criteria
    if abs(x_new - x) < tolerance:
        break

    # Update the guess
    x = x_new

# Plotting the results
plt.plot(iterations, approximations, marker='o')
plt.axhline(y=np.sqrt(2), color='r', linestyle='--', label=f"sqrt(2) ≈ {np.sqrt(2):.7f}")
plt.title("Newton's Method Approximations of sqrt(2)")
plt.xlabel("Iteration")
plt.ylabel("Iterate value")
plt.legend()
plt.grid(True)
plt.show()

plt.plot(iterations, np.array(approximations) - np.sqrt(2), marker='o')
plt.xlabel("Iteration")
plt.ylabel("Error")
plt.yscale("log")

plt.grid(True)
plt.show()

print("\n sqrt(2) ≈ ", x_new)
```
